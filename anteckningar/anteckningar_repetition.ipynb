* Docker
    - docker gr det enklare att f projekt/program att fungera p andra datorer
    - Dockerfile -> docker image -> docker container
        - dockerfile: innehller alla instruktioner/installationer. En Dockerfil r en receptlista som beskriver stegen fr att packa ihop allt som behvs fr en programvara i en Docker-bild.
        - docker image: En Docker-bild r som en frdigpackad lda med allt en programvara behver fr att kra. Det inkluderar koden, verktygen och instllningarna. Du kan dela och kra dessa ldor verallt med hjlp av Docker.
        - docker container: kra ditt program. En Docker-container r som en isolerad mini-dator dr en frdigpackad programvara kan kras utan att pverka resten av systemet.

* docker-compose.yaml
    - kra flera docker containers
    - lgger allt i samma vrld, alla containers, s att dem enkelt kan kommunicera
        - dash board frn python, data frn en databas
        - dem kan enkelt kommunicera med varandra med docker-compose
        - olika containers har ansvar fr olika saker, hela projektet ska vara modulrt
