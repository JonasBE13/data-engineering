* Docker
    - docker gör det enklare att få projekt/program att fungera på andra datorer
    - Dockerfile -> docker image -> docker container
        - dockerfile: innehåller alla instruktioner/installationer. En Dockerfil är en receptlista som beskriver stegen för att packa ihop allt som behövs för en programvara i en Docker-bild.
        - docker image: En Docker-bild är som en färdigpackad låda med allt en programvara behöver för att köra. Det inkluderar koden, verktygen och inställningarna. Du kan dela och köra dessa lådor överallt med hjälp av Docker.
        - docker container: köra ditt program. En Docker-container är som en isolerad mini-dator där en färdigpackad programvara kan köras utan att påverka resten av systemet.

* docker-compose.yaml
    - köra flera docker containers
    - lägger allt i samma värld, alla containers, så att dem enkelt kan kommunicera
        - dash board från python, data från en databas
        - dem kan enkelt kommunicera med varandra med docker-compose
        - olika containers har ansvar för olika saker, hela projektet ska vara modulärt
